```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dataset_all.csv')
data.head()
```

```python
# Statistical Ti, V, Fe, Co, Ni, Zr, Mo, Sn, W column the number of true
Ti = data['Ti'].value_counts().get(True, 0)
V = data['V'].value_counts().get(True, 0)
Fe = data['Fe'].value_counts().get(True, 0)
Co = data['Co'].value_counts().get(True, 0)
Ni = data['Ni'].value_counts().get(True, 0)
Zr = data['Zr'].value_counts().get(True, 0)
Mo = data['Mo'].value_counts().get(True, 0)
Sn = data['Sn'].value_counts().get(True, 0)
W = data['W'].value_counts().get(True, 0)

# Draw a bar chart of the number of true values of each element
# Combine the counts into a single DataFrame
elements_counts = pd.DataFrame({
    'Element': ['Ti', 'V', 'Fe', 'Co', 'Ni', 'Zr', 'Mo', 'Sn', 'W'],
    'Count': [Ti, V, Fe, Co, Ni, Zr, Mo, Sn, W]
})

# Plot the bar chart with larger font size
plt.figure(figsize=(12, 9), dpi=600)
barplot = sns.barplot(x='Element', y='Count', data=elements_counts)
plt.title('Number of True Values for Each Element', fontsize=20)
plt.xlabel('Element', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

for index, row in elements_counts.iterrows():
    barplot.text(row.name, row.Count, row.Count, color='black', ha="center", fontsize=12)

plt.show()
```

```python
sns.set_style('whitegrid')
sns.set_context("notebook", font_scale=1.5)  # è°ƒæ•´å­—ä½“å¤§å°
plt.figure(figsize=(16, 12), dpi=600)

# ç»˜åˆ¶ CM_type çš„é¢‘çŽ‡åˆ†å¸ƒå›¾ï¼ŒæŒ‰é™åºæŽ’åˆ—
plt.subplot(2, 2, 1)
sns.countplot(data['CM_type'], order=data['CM_type'].value_counts().index)
plt.title('CM_type Frequency Distribution')

# ç»˜åˆ¶ CM_morph çš„é¢‘çŽ‡åˆ†å¸ƒå›¾ï¼ŒæŒ‰é™åºæŽ’åˆ—
plt.subplot(2, 2, 2)
sns.countplot(data['CM_morph'], order=data['CM_morph'].value_counts().index)
plt.title('CM_morph Frequency Distribution')

# ç»˜åˆ¶ MS2_morph çš„é¢‘çŽ‡åˆ†å¸ƒå›¾ï¼ŒæŒ‰é™åºæŽ’åˆ—
plt.subplot(2, 2, 3)
sns.countplot(data['MS2_morph'], order=data['MS2_morph'].value_counts().index)
plt.title('MS2_morph Frequency Distribution')

# ç»˜åˆ¶ CP_morph çš„é¢‘çŽ‡åˆ†å¸ƒå›¾ï¼ŒæŒ‰é™åºæŽ’åˆ—
plt.subplot(2, 2, 4)
sns.countplot(data['CP_morph'], order=data['CP_morph'].value_counts().index)
plt.title('CP_morph Frequency Distribution')

plt.tight_layout()
plt.show()
```
